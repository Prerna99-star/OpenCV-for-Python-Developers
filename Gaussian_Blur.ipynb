```
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```
image = cv2.imread("thresh.jpg", 1)
plt.imshow(image)
```

```
blur = cv2.GaussianBlur(image, (5,55), 0)  #keeping the odd values of blur, doing more blur at y axis
plt.imshow(blur)
```
