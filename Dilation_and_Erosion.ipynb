```
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```
image = cv2.imread("thresh.jpg", 1)
plt.imshow(image)
```

```
kernel = np.ones((5,5), 'uint8')
```

```
dilate = cv2.dilate(image, kernel, iterations = 1)
erode = cv2.erode(image, kernel, iterations = 1)
```

```
plt.imshow(dilate)
```

```
plt.imshow(erode)
```
